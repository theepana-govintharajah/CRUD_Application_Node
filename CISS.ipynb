```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['AIT302'], label='Time Series Data', color='blue')

# Customize the plot
plt.title('Time Series Plot')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.grid(True)

# Format the x-axis labels for better readability (optional)
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(prune='both'))

# Show a legend (optional)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
